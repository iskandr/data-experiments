```python
import numpy as np
import pylab 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm
from matplotlib import cm

class Rosenbrock(object):
    def __init__(self, a = 100, b=1, perturb = False, perturb_scale=10, ndims = 2):
        if perturb:
            a += np.random.randn() * perturb_scale / ndims
            b += 0.01 * np.random.randn()
        self.a = a 
        self.b = b 
        self.ndims = ndims
        self.perturb_scale = perturb_scale
        
    def __call__(self, x):
        a,b = self.a, self.b
        x,y = x[:-1], x[1:]
        return np.sum(a * (y-x**2)**2 + b * (1-x)**2)
    
    def perturb(self):
        return Rosenbrock(a = self.a, b = self.b, perturb_scale = self.perturb_scale, ndims = self.ndims, perturb = True)
    
    def grad(self, x):
        """
        From Matlab:
        df = zeros(D, 1);
        df(1:D-1) = - 400*x(1:D-1).*(x(2:D)-x(1:D-1).^2) - 2*(1-x(1:D-1));
        df(2:D) = df(2:D) + 200*(x(2:D)-x(1:D-1).^2);
        """
        x,y = x[:-1], x[1:]
        a,b = self.a, self.b
        d = np.zeros(self.ndims)
        
        d[:-1] = -4 * a* x* (y - x **2) - 2*b*(1-x)
        d[1:] += 2*a *(y-x**2)
        
        return d


    
    def hessian(self, x):
        """
        From Matlab:
          ddf = zeros(D,D);
          ddf(1:D-1,1:D-1) = diag(-400*x(2:D) + 1200*x(1:D-1).^2 + 2);
          ddf(2:D,2:D) = ddf(2:D,2:D) + 200*eye(D-1);
          ddf = ddf - diag(400*x(1:D-1),1) - diag(400*x(1:D-1),-1);
        """
        
        a,b = self.a, self.b
        D = self.ndims 
        
        h = np.zeros ( (D,D) )
        h = np.diag(-4*a*x[:-1],1) + np.diag(-4*a*x[:-1],-1)
        h[:D-1, :D-1] += np.diag(-4 * a * x[1:] + 12*a*x[:-1]**2.0 + 2*b)
        h[1:, 1:] += 2*a*np.eye(D-1)                      
        
        if self.ndims == 2:
          # 2 dimensional sanity check against the old code 
          x,y = x[0], x[1]
          dxx = -4 * a * (y-3*x**2) + 2 * b # == -4ay + 12ax^2 + 2b
          dxy = dyx = -4 * a * x 
          dyy = 2 * a 
          h_small = np.array([[dxx, dxy],[dyx, dyy]])
          assert ((h - h_small) ** 2).mean() < 0.001, "generalized %s != %s" % (h, h_small)
        return h
    
    def plot(self, azim = 115, elev=45):
      fig = plt.figure()
      ax = Axes3D(fig, azim = azim, elev =elev)

      s = .05
      X = np.arange(-0.3, 1.3+s, s) 
      Y = np.arange(-0.3, 1.3+s, s)
      X, Y = np.meshgrid(X, Y)   
      Z = self.a*(Y-X*X)**2 + self.b * (1.-X)**2  # rosenbrock function
      ax.plot_surface(X, Y, Z, cstride = 2, rstride = 2, norm = LogNorm(), cmap = cm.hot) 
      return ax 
        
    def plot_trajectory(self, xs, marker = 'b+', markersize = 12, azim = 115, elev = 45):
      """
      Plot a trajectory of iterates over the Rosenbock function
      """  
      ax = self.plot( azim = azim, elev = elev)
      ax.plot([x[0] for x in xs], [x[1] for x in xs], marker, markersize=markersize, markeredgewidth=3, markeredgecolor='b')
      return ax
    
```

```python
class Optimizer(object):
  def __init__(self, f, niters = 3000, 
               step_sizes = None, 
               max_step_size_exp = 1, 
               min_step_size_exp = -5, 
               step_size_exp_increment = 0.25, 
               momentum = 0, 
               min_decrease_ratio = 0.9999, 
               max_increases = 1, 
               perturbed_line_search = False,
               line_search = True, 
               verbose = False):
      self.f = f 
      self.niters = niters 
      if step_sizes is None:
        step_sizes = 10.0**np.arange( min_step_size_exp, max_step_size_exp, step_size_exp_increment)
      self.step_sizes = step_sizes 
      self.momentum = momentum
      self.min_decrease_ratio = min_decrease_ratio
      self.max_increases = max_increases 
      self.verbose = verbose 
      self.perturbed_line_search = perturbed_line_search 
      self.line_search = line_search
  
  def __str__(self):
    fields = ", ".join(["%s = %s" % (k,v) for (k,v) in self.__dict__.items() 
                        if not k.startswith("_") and k != 'step_sizes' and k != 'f'])
    return "%s(%s)" % (self.__class__.__name__, fields)
      

  
  def run(self, x = None, plot = False, full_history = False, target = None):
    if x is None:
      x = np.random.randn(self.f.ndims)
    else:
      x = x
    x = np.asarray(x)
    last_step = np.zeros(self.f.ndims)
    xs = []
    values = []
    n_increases = 0
    for curr_iter in xrange(self.niters):
        # algorithms with local/nested optimization routines might need to 
        # know which epoch we're on, i.e. for learning rate decay 
        self.curr_iter = curr_iter
        fx = self.f(x)
        if target and fx < target:
            break
        xs.append(x.copy())
        values.append(fx)
        
        if self.verbose:
            print "Iter", epoch, "x =", x, "dx =", dx 
        
        opt_result = self.search_direction(x)
        if isinstance(opt_result, tuple):
            dx, x  = opt_result
        else:
            dx = opt_result 
        dx += self.momentum * last_step
        
        if self.line_search:
            
            best_x = x 
            best_dx = np.zeros_like(x)
            best_value = np.inf
            best_step_size = None
            
            if self.perturbed_line_search:
                f = self.f.perturb()
            else:
                f = self.f 
            for step_size in self.step_sizes:
                candidate_dx = step_size * dx 
                candidate_x = x + candidate_dx 
                candidate_value = f(candidate_x)
                
                if candidate_value < best_value:
                    best_x = candidate_x
                    best_dx = candidate_dx
                    best_value = candidate_value
                    best_step_size = step_size
            if self.perturbed_line_search:
                best_value = self.f(best_x)
        else:
            
            best_dx = 0.005 * dx
            best_x = x + best_dx
            best_value = self.f(best_x)
        
        value_ratio = (best_value / fx)
        if curr_iter % 100 == 0:
            print curr_iter, best_value, value_ratio, best_step_size 
        if value_ratio > self.min_decrease_ratio:
            n_increases += 1
        if n_increases > self.max_increases:
          if self.verbose: 
            print "No improvement from any step sizes at iteration", i
          break 
        if self.verbose: 
            print "---", "step size", best_step_size
            print "---", "new function value", best_value 
        x = best_x
        last_step = best_dx 
    if plot:
        self.f.plot_trajectory(xs)
    if full_history:
        return curr_iter, xs, values, best_value < target 
    else:
        return curr_iter, best_value, best_value < target 

class GradientDescent(Optimizer):
  def search_direction(self,x):
        return -self.f.grad(x)

def newton_step(f, x):
    dx = f.grad(x)
    h = f.hessian(x)
    p = np.linalg.solve(h, -dx)
    return p
        
class Newton(Optimizer):
    def search_direction(self, x):
        return newton_step(self.f, x)

class BFGS(Optimizer):
    def run(self, *args, **kwargs):
        # initialize estimate of inverse hessian matrix
        self.hinv = np.eye(self.f.ndims)
        self.last_x = np.zeros(self.f.ndims)
        self.last_dx = np.zeros(self.f.ndims )
        return Optimizer.run(self, *args, **kwargs)
   
        
    def search_direction(self, x):
        dx = self.f.grad(x)
        s = x - self.last_x
        y = dx - self.last_dx 
        
        sy  = np.dot(s,y)
        rho = 1 / sy
        left = np.eye(self.f.ndims) - rho * np.outer(s,y)
        right = np.eye(self.f.ndims) - rho * np.outer(y,s)
        SS = np.outer(s,s)
        self.hinv = np.dot(np.dot(left, self.hinv), right) + rho * SS
        
        self.last_x = x 
        self.last_dx = dx 
        return -np.dot(self.hinv, dx)
        
def default_arg(kwargs,k,default):
    if k in kwargs:
      v = kwargs[k]
      del kwargs[k]
    else:
      v = default
    return v  
class Dist(Optimizer):
    def __init__(self, *args, **kwargs):
        self.n_workers = default_arg(kwargs, 'n_workers', 2)
        if 'perturbed_line_search' not in kwargs:
          kwargs['perturbed_line_search'] = True
        Optimizer.__init__(self, *args, **kwargs)
        
    def perturbed_function(self):
        return Rosenbrock(self.f.a, self.f.b, perturb=True, ndims = self.f.ndims) 
    
    def perturbed_functions(self):
        return [self.perturbed_function() for _ in xrange(self.n_workers)]
    
class AvgGradientDescent(Dist):
    def search_direction(self, x):
        grads = [f.grad(x) for f in self.perturbed_functions()]
        combined = np.mean(grads, axis=0)
        return -combined

class AvgNewton(Dist):
    def search_direction(self, x):
        dxs = [newton_step(f, x) for f in self.perturbed_functions()]
        return np.mean(dxs, axis=0)

class BestGradient(Dist):
    def search_direction(self, x):
        held_out = self.perturbed_function()
        best_value = np.inf 
        best_dx = None
        for dx in [-f.grad(x) for f in self.perturbed_functions()]:
            v = held_out(x+0.01*dx)
            if v < best_value:
                best_value = v 
                best_dx = dx
        return best_dx


import random 

class DistNewton(Dist):
    """
    Estimate the Hessian by getting (x-x_0) and (dx - dx_0) 
    constraints from each worker
    """
  
    
    def __init__(self, *args, **kwargs):
        self.n_local_updates = default_arg(kwargs, 'n_local_updates', 10)
        self.local_step_size = default_arg(kwargs, 'local_step_size', 10.0**(-4))
        self.local_step_decays = default_arg(kwargs, 'local_step_decays', True)
        self.local_momentum = default_arg(kwargs, 'local_momentum', 0)
        self.use_newer_gradients = default_arg(kwargs, 'use_newer_gradients', False)
        self.random_sampling = default_arg(kwargs, 'random_sampling', False)
        # options:
        # 'sr1' : build hessian out of n_workers symmetric rank-1 updates
        # 'bfgs' : build inverse hessian out of n_workers BFGS updates 
        # 'l-bfgs' : compute Hinv*g directly without storing Hinv
        # 'l-bfgs-avg' : just a rank-2 Hinv over averaged change in x and gradient
        # 'diagonal' : get the median ddx/dx ratio from each worker 
        # 'finite-difference' : move along gradient, compute new gradient, take diff as estimate of Hg 
        # 'global-bfgs' : do a global BFGS, preserving inverse hessian between epochs
        # 'global-rank2' : constract a single rank-2 hessian for each iteration by using
        #           average change-in-x, change-in-dx vectors
        # 'variance' : rescale each dimension to give it unit variance in the noise
        #              of per-worker changes in x 
        # 'lstsq' : least squares approximation of Hessian
        self.method = default_arg(kwargs, 'method', 'l-bfgs') 
        
        Dist.__init__(self, *args, **kwargs)
        
    
    def search_direction(self, x):
        old_grads = []
        start_xs = []
        new_xs = []
        new_grads = []
        old_values = []
        new_values = []
        
        for (worker_idx, perturbed_fn) in enumerate(self.perturbed_functions()):
          if self.local_step_size == 'random':
            step_size = 10**-np.random.uniform(2.5, 6)
          else:
            step_size = self.local_step_size 
          if self.local_step_decays:
            step_size /= (1+self.curr_iter) 
        
          start_x = x.copy()
          start_xs.append(start_x)       
          old_grads.append(perturbed_fn.grad(start_x))
          old_values.append(perturbed_fn(start_x))
          local_x = start_x.copy()
          last_step_dir = np.zeros(self.f.ndims)
          for local_update_idx in xrange(self.n_local_updates):
              
              local_grad = perturbed_fn.grad(local_x)
              assert not np.any(np.isnan(local_grad)), (self.curr_iter, local_update_idx)
              gn = np.linalg.norm(local_grad)
              
              assert gn < 1000000
              momentum =  self.local_momentum 
              step_dir = -step_size * local_grad + momentum * last_step_dir
              local_x += step_dir 
              if self.random_sampling:
                local_x += 1/(self.f.ndims) * np.random.randn(len(step_dir)) * np.sqrt(np.abs(step_dir))  
              last_step_dir = step_dir 
          #print worker_idx, perturbed_fn(local_x) -perturbed_fn(start_x) 
          new_xs.append(local_x)
          new_grads.append(perturbed_fn.grad(local_x))
          new_values.append(perturbed_fn(local_x))

        # want to solve for H s.t. H(x-x0) = -(dx-dx0)
        # but across multiple perturbed s_i = (x_i - x0), y_i = (dx_i - dx0)
        # creating a linear system H * [s_1, ... s_k] = -[y_1, ..., y_k]
        # where k is the number of workers 
        
        S = np.array([new_x - start_x for (new_x,start_x) in zip(new_xs, start_xs)])
        Y = np.array([new_grad-old_grad 
                      for (new_grad,old_grad) 
                      in zip(new_grads,old_grads)])
          
        Yavg = np.mean(Y, axis = 0)
        Savg = np.mean(S, axis = 0)
       
        old_Vavg = np.mean(old_values)
        V = np.array(new_values)
        Vavg = np.mean(V)
        
        if self.n_workers == 1:
            assert np.all( (Yavg == Y[0]) | np.isnan(Yavg)), "Yavg %s, Y %s, new_grad %s" % (Yavg, Y, new_grad)
            assert np.all( (Savg == S[0]) | np.isnan(Savg)), "Savg %s, S %s, new_x %s" % (Savg, S, new_x)
        I = np.eye(self.f.ndims)
        
        old_grad = np.mean(old_grads, axis = 0)  
        new_grad = np.mean(new_grads, axis = 0)
        if self.use_newer_gradients:
          g = new_grad
        else:
          g = old_grad
        
            
       
        def bfgs_update(s,y, B = I):
           Bs = np.dot(B, s)
           top = np.dot(Bs, Bs.T)
           bottom = np.dot(s, Bs)
           return B + np.outer(y,y,) / np.dot(y,s) - top/bottom 
         
        def dfp_rank2(s,y):
          rho = np.dot(y,s)
          left = I - np.outer(y,s)/rho
          right = I - np.outer(s,y)/rho
          return  np.dot(left, right) + np.outer(y,y)/rho
      
        def bfgs_update_inv(s, y, Hinv = I):
          sy  = np.dot(s,y)
          rho = 1 / sy
          left = np.eye(len(s)) - rho * np.outer(s,y)
          right = np.eye(len(s)) - rho * np.outer(y,s)
          SS = np.outer(s,s)
          return np.dot(np.dot(left, Hinv), right) + rho * SS 
        
        def sr1_update(s,y, B = I):
          Bs = np.dot(B,s)
          diff = y - Bs
          rank1 = np.outer(diff, diff)
          denom = np.dot(diff, s)
          return B + rank1 / denom 
        
        def lbfgs(S = S, Y = Y, g=g):
          q = g
          rhos = [] 
          alphas = []
          for (s,y) in zip(S,Y):
             rho = 1 / np.dot(s,y)
             rhos.append(rho)
             alpha = rho * np.dot(s, q)
             alphas.append(alpha)
             q -= alpha * y
          r = q # Should we be multiplying by some diagonal inverse hessian here?
          for i, (s,y) in reversed(list(enumerate(zip(S,Y)))):
             rho = rhos[i]
             alpha = alphas[i]
             beta = rho * np.dot(y,r)
             print rho, alpha, beta 
             r += s * (alpha - beta)
          return -r 
        
          
        if self.method =='global-bfgs':
          if hasattr(self, 'h_inv'):
            h_inv = self.h_inv 
          else: 
            h_inv = I 
          h_inv = bfgs_update_inv(Savg, Yavg, h_inv)
          self.h_inv = h_inv 
          search_dir = -np.dot(h_inv, g)
        elif self.method =='global-rank2':
          # rank-2 hessian ... maybe a bad idea?   
          h_approx = dfp_rank2(Savg, Yavg)
          search_dir = -np.linalg.solve(h_approx, g)  
        elif self.method == 'bfgs':
          h_inv = I
          for i in xrange(self.n_workers):
            h_inv = bfgs_update_inv(S[i,:],Y[i,:],h_inv)
          search_dir = -np.dot(h_inv, g) 
        elif self.method == 'sr1': 
          h_approx = I
          for (s,y) in  zip(S,Y):
            h_approx += sr1_update(s,y,B=h_approx)
          sr1_averaging = False
          if sr1_averaging:
            h_approx /= self.n_workers 
          search_dir = -np.linalg.solve(h_approx, g)  
            
        elif self.method == 'diag':
          # change in gradient per change in position at each worker 
          ratios = S/Y 
          assert ratios.shape == (self.n_workers, self.f.ndims) 
          h_inv_diag = np.mean(ratios, 0)
          search_dir = -g * h_inv_diag
            
 
        elif self.method == 'l-bfgs':
         search_dir = lbfgs(S,Y,g) 
            
        elif self.method == 'l-bfgs-avg':
         search_dir = lbfgs([Savg], [Yavg], g)
        
        elif self.method == 'fn-min':
          s = np.std(new_values)
          perturbed = self.f.perturb()
          values = np.array([perturbed(new_x) for new_x in new_xs])
          s = np.std(values)
          m = np.mean(values)
          diffs = [ -(v - m)/s for v in values]
          weights = np.exp(diffs)
          w = np.mean([s*wi for s,wi in zip(S,weights)], axis=0)
          w /= np.sum(weights)
          search_dir = lbfgs(S,Y,-w/np.linalg.norm(w))
            
        elif self.method == 'norm-min':
          w = np.linalg.lstsq(Y, [np.linalg.norm(ng) for ng in new_grads])[0]
          search_dir = lbfgs(S,Y,w/np.linalg.norm(w))
            
        elif self.method == 'norm-min2':
          w = np.linalg.lstsq(S, [np.linalg.norm(ng) for ng in new_grads])[0]
          search_dir = lbfgs(S,Y, -w / np.linalg.norm(w) * np.linalg.norm(g) + g)
        
        elif self.method == 'svd':
            U, D, V = np.linalg.svd(Y.T, full_matrices=False)
            k = np.argmax(D / D[0] < 10**-6)
            U = U[:, :k]
            D = D[:k]
            V = V[:k, :]
            Ug = np.dot(U.T, g)
            
            DinvUg = np.dot(np.diag(1.0 / D), Ug)
            VDinvUg = np.dot(V.T, DinvUg)
            search_dir = -np.dot(S.T, VDinvUg)
        else:        
          assert self.method == 'lstsq', "Unrecognized algorithm for building Hessian: %s" % self.hessian
          H = np.linalg.lstsq(S, Y, rcond=0.0001)[0].T
          # symmetrize?
          H = (H + H.T) * 0.5   
          search_dir = -np.linalg.solve(H, g)
        if self.use_newer_gradients:
            return search_dir, np.mean(new_xs, axis=0)
        else:
            return search_dir
 
class KeepBest(object):
    def __init__(self):
        self.value = np.inf
        self.converged = False 
        self.iters = 10000000
        self.key = None 
    def __str__(self):
        return "%s, converged = %s, iters = %s, value = %s" % (self.key, self.converged, self.iters, self.value)
        
    def update(self, key, result):
        i, v, c = result 
        print key
        print "-- %s after %d iters (converged = %s)" % (v, i, c)
        print 
        
        keep = False
        if c and not self.converged:
            keep = True
        elif not c and self.converged:
            keep = False
        elif c:
            keep = i < self.iters 
        elif not c:
            keep = v < self.value or (v == self.value and i < self.iters)
            
        if keep:
            self.value = v
            self.iters = i
            self.key = key
            self.converged = c
```
